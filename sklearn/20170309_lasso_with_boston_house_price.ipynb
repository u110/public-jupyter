```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
pd.DataFrame(boston.target, columns=["houseprice"]).plot.hist()
```

```python
from sklearn.linear_model import MultiTaskLasso, Lasso
```

```python
alpha_arr = np.arange(0.1, 3, 0.1)

for alpha_ in alpha_arr:
    coef_lasso_ = np.array([Lasso(alpha=alpha_).fit(boston.data, boston.target).coef_ ])
    plt.spy(coef_lasso_)
    plt.show()
```
