```python
import pandas as pd
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2016", periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000,4), 
                  index=ts.index, columns=["aa","bb","Cc","dD"])
df = df.cumsum()
df.plot();
```

```python
df.ix[5].plot(kind='bar');plt.axhline(0, color='k')
```

```python
df2 = pd.DataFrame(np.random.rand(10, 4), columns=['a','b','c','d'])
df2.plot.bar();
```

```python
df2.plot.bar(stacked=True);
```

```python
df2.plot.barh(stacked=True)
```

## histogram

```python
df4 = pd.DataFrame({
    'a': np.random.randn(1000) + 1, 
    'b': np.random.randn(1000),
    'c': np.random.randn(1000) - 1,
    },
    columns = ['a', 'b', 'c']        
)
df4.plot.hist(alpha=.5)
```

```python
df4.plot.hist(stacked=True, bins=20)
```

```python
df4['a'].plot.hist(orientation='horizontal', cumulative=True)
```

## subplot

```python
df4.hist(alpha=.5, bins=50)
```

```python
data = pd.Series(np.random.randn(1000))
data.hist(by=np.random.randint(0,9,1000), figsize=(6,6))
```

## box plot

```python
df = pd.DataFrame(np.random.rand(10, 5), columns=list("ABCDE"))
df.plot.box()
```

```python
color = dict(boxes='DarkGreen', whiskers='DarkOrange',
            medians='DarkBlue', caps='Grey')
df.plot.box(color=color, sym='r+')
```

```python
df.plot.box(vert=False, positions=[5,4,3,2,1])
```

```python
df.boxplot()
```

```python
df = pd.DataFrame(np.random.rand(10,2), columns=['Col1', 'Col2'])
df['X'] = pd.Series(list('AAAAABBBBB'))
df.boxplot(by='X')
```
