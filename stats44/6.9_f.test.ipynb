```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
def T(sigma2_a, sigma2_b):
    if sigma2_a > sigma2_b:
        return sigma2_a / sigma2_b
    else:
        return sigma2_b / sigma2_a
```

```python
def f_test(a, b, size_a, size_b):
    f_value = T(a, b)
    return stats.f.cdf(f_value, size_a, size_a), f_value
```

```python
a = 2.344
b = 1.068
n = 50
df = n - 1
```

```python
f_test(a, b, 50, 50)
```

```python
alpha = 0.05
upr = (1-alpha/2) 
lwr = alpha/2 
lwr, upr, stats.f.ppf([lwr, upr], df, df)
```

- H_0:  a = b
- H_1: a != b


T = 2.195 > F(49,49; 0.025)=1.762 なので H_0 は棄却

```python
x = np.linspace(stats.f.ppf(0.002, df, df), stats.f.ppf(0.99, df, df), 100)
plt.plot(x, stats.f.pdf(x, df, df))
plt.axvline(x=stats.f.ppf([lwr, upr], df, df)[0])
plt.axvline(x=stats.f.ppf([lwr, upr], df, df)[1])
```

```python
plt.plot(x, stats.f.cdf(x, df, df))
```
